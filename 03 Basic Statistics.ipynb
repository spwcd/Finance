# Variance ความแปรปรวนของข้อมูล

สมมุติว่ามีระบบสองระบ

ระบบ A

ปีที่ 1 21%

ปีที่ 2 19%

ปีที่ 3 22%

ปีที่ 4 18%

ปีที่ 5 20%


ระบบ B

ปีที่ 1 55%

ปีที่ 2 -15%

ปีที่ 3 -25%

ปีที่ 4 55%

ปีที่ 5 30%

```python
system_A = (21,19,22,18,20)
system_B = (55,-15,-25,55,30)

plt.plot(system_A)
plt.plot(system_B)
```

```python
plt.plot(np.cumsum(system_A))
plt.plot(np.cumsum(system_B))
```

```python
print('mean of system A:',np.mean(system_A))
print('mean of system B:',np.mean(system_B))
```

Variance equation

<img src="Pics/var.png">

```python
var_A = np.sum(((system_A - np.mean(system_A))**2))/len(system_A)
var_A
```

```python
var_B = np.sum(((system_B - np.mean(system_B))**2))/len(system_B)
var_B
```

```python
print('Var of system A:',np.var(system_A))
print('Var of system B:',np.var(system_B))
```

# Standard Deviation

Variance equation

<img src="Pics/std.png">

```python
print('Std of system A:', var_A**0.5)
print('Std of system B:', var_B**0.5)
```

```python
print('Var of system A:',np.std(system_A))
print('Var of system B:',np.std(system_B))
```

# Covariance หุ้นสองตัวมันขึ้นลงตามกันหรือไม่

```python
study_hr = (15,4,12,6,10)
grade = (3.5,1,3,2,4)
plt.scatter(study_hr,grade,s=500)
```

Covariance equation

<img src="Pics/cov.png">

```python
(np.sum(((study_hr - np.mean(study_hr)))*((grade - np.mean(grade)))))/(len(study_hr)-1)
```

```python
np.cov(study_hr,grade)
```

# Correlation หุ้นสองตัวมันขึ้นลงตามแค่ไหน

Correlation equation

<img src="Pics/corr.png">

```python
((1/5)*np.sum(((study_hr - np.mean(study_hr)))*((grade - np.mean(grade)))))/(np.std(study_hr)*np.std(grade))
```

```python
np.corrcoef(study_hr,grade)
```
