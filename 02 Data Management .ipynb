# Lead/Lag

```python
SET['Close-lag1'] = SET['Close'].shift(1)
SET['Close-lag2'] = SET['Close'].shift(2)
SET[['Close', 'Close-lag1']].head()
```

```python
SET['Close-lead1'] = SET['Close'].shift(-1)
SET['Close-lead2'] = SET['Close'].shift(-2)
SET[['Close', 'Close-lag1', 'Close-lead1']].head()
```

```python
SET[['Close', 'Close-lag1', 'Close-lead1', 'Close-lag2', 'Close-lead2']].plot()
```

```python
SET[['Close', 'Close-lag1', 'Close-lead1', 'Close-lag2', 'Close-lead2']]['2018':].plot(linewidth=2)
```

```python
SET[['Close', 'Close-lag1', 'Close-lead1']]['2018':].plot(linewidth=2)
```

# Plot data with volume

# Load Multiple stock

```python
stocklist = ['HMPRO', 'PTT', 'CPF', 'TMB','STA']
port = pd.DataFrame()
for i in range(0,len(stocklist)):
    stock = getStock(stocklist[i], allStocks)
    port[i] = stock['Close']

port.columns = stocklist
port.head()
```

# Missing Value

```python
port = port[:'2018-03']
port.plot()
```

```python
port[:100].plot()
```

```python
port.dropna(inplace=True)
port.head()
```

```python
port[:100].plot()
```

# Normalize

Normalize
$$
\frac {P_t}{P_0} * 100
$$

```python
portNorm = (port/port.iloc[0]) *100
portNorm.columns = stocklist
portNorm.head()
```
